# Creating dummy data
When experimenting with programs related to machine learning, there are times when you want to have random data (dummy data) that is not real data. Let's get familiar with NumPy by creating dummy data with random numbers that follow a normal distribution.


We will create data that looks like the following when drawn in a scatter plot

## Problem 1: Creating random numbers
Create 500 random numbers with a two-dimensional normal distribution whose mean is (-3, 0) and whose covariance matrix is 
$\left[\begin{array}{ll}1.0 & 0.8 \\ 0.8 & 1.0\end{array}\right]$

Multidimensional normally distributed random numbers can be created by using np.random.multivariate_normal(). The shape of the created ndarray will be (500, 2). To ensure that the value does not change from one run to the next, we will fix the seed value of the random number using np.random.seed() . We will use a seed value of 0.



```python
np.random.seed(0)
sample = np.ones((500,2))
cov = np.array([1.0,0.8,0.8,1.0]).reshape(2,2)
for z in range(1,500):
    sample[z] = np.random.multivariate_normal([-3,0],cov)
```

## Problem 2: Visualization by scatter plot
Visualize the data points created in Problem 1 with a scatter plot.
Scatter plots can be drawn by using matplotlib's plt.scatter() .

```python
rng = np.random.RandomState(0)
colors = rng.rand(500)
_ = plt.scatter(sample[:,0],sample[:,1],c=colors)
```

## Problem 3: Visualization by histogram
Visualize the data created in Problem 1 with a histogram.
A histogram can be drawn by using plt.hist() .

Create it for each dimension and create two histograms.
The range of the horizontal axis of each histogram should be aligned using plt.xlim() .

```python
plt.title("Histogram X axis")
plt.xlabel("x")
plt.ylabel("Frequency")
plt.xlim(-6,2)
_ = plt.hist(sample[:,0],bins=50)
```

```python
plt.title("Histogram Y axis")
plt.xlabel("y")
plt.ylabel("Frequency")
plt.xlim(-4,3)
_ = plt.hist(sample[:,1],bins=50)
```

## Problem 4: Addition of data
Create 500 new random numbers with a two-dimensional normal distribution whose mean is (0, -3) and whose covariance matrix is represented by the following.

$\left[\begin{array}{ll}1.0 & 0.8 \\ 0.8 & 1.0\end{array}\right]$

Then, visualize the data of problems 1 and 4 as one scatter plot. As a legend, show 0for Problem 1 and1for Problem 4.

```python
np.random.seed(0)
sample_additional = np.ones((500,2))
cov_additional = np.array([1.0,0.8,0.8,1.0]).reshape(2,2)
for z in range(1,500):
    sample_additional[z] = np.random.multivariate_normal([0,-3],cov)
```

```python
plt.scatter(sample[:,0],sample[:,1],c="blue",label="Original Data")
plt.scatter(sample_additional[:,0],sample_additional[:,1],c="orange",label="Additional Data")
_ = plt.legend()
```

## Problem 5: Data combination
It is better to keep the data together so that it is easier to deal with later. Combine the ndarrays created in Problems 1 and 4 intoa single ndarray of (1000, 2). The concatenation can be done using np.concatenate() or np.vstack() .

```python
concat_data = np.concatenate((sample,sample_additional),axis=0)
concat_data.shape
```

## Problem 6: Labeling
For each of the 1000 pieces of data that you have combined together, label them to indicate whether they are from question 1 or 4. Question 1 should be labeled 0 and Question 4 should be labeled 1 .


Create a ndarray of (1000, 3) by adding a new column of labels containing 0s and 1s.


Datasets used for machine learning often have this form.

```python
label_data = np.zeros((1000,3))
label_data[:,0:2] = concat_data[:,0:2]
label_data[0:499,2] = 0
label_data[500:1000,2] = 1
```
