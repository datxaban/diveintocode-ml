# Fisher's iris
You will learn how to handle basic data using the world-famous data set of machine learning and data analysis called "Fisher's Iris".


This is data from the following three types of irises, totaling 150, collected on the same day in a pasture on the Gaspé Peninsula, Canada.

<li>Iris setosa --Wikipedia

Setosa, virgiclor, and virginica are what are called objective variables or labels here.​


Four types of information are recorded for each plant: sepal_length (length of sepal), sepal_width (width of sepal), petal_length (length of petal), and petal_width (width of petal). These four types are called feature values.

## Problem 6: Creating a diagram
It is important to visualize and confirm the data. Do the following three things. In addition to Matplotlib, seaborn is also often used as a drawing library.


<li>Make a pie chart of the number of samples per label (while showing percentages)
<li>Select features one by one and visualize the distribution of data for each label using a box plot
<li>Select features one by one and visualize the distribution of data for each label using a violin plot



```python
labels = 'Setosa', 'Virgicolor', 'Virginica'
sizes = [data[data["Species"] == 0]["Species"].count(), data[data["Species"] == 1]["Species"].count(), data[data["Species"] == 2]["Species"].count()]
explode = (0, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```

```python
labels = ['sl', 'sw', 'pl','pw']

fig, (ax1, ax2,ax3) = plt.subplots(nrows=1, ncols=3, figsize=(9, 4))

# rectangular box plot
bplot1 = ax1.boxplot(data[data["Species"] == 0].iloc[:,0:4],
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Iris setosa')

bplot2 = ax2.boxplot(data[data["Species"] == 1].iloc[:,0:4],
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('Iris virgicolor')

bplot3 = ax3.boxplot(data[data["Species"] == 2].iloc[:,0:4],
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax3.set_title('Iris virginica')
# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2, bplot3):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
   

# adding horizontal grid lines
for ax in [ax1, ax2, ax3]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Widths and Lengths')
    ax.set_ylabel('feature  values')

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

axs[0].violinplot(data[data["Species"] == 0].iloc[:,0:4], showmeans=True, showextrema=True, showmedians=True)
axs[0].set_title('Iris setosa')
# ax[0].set_xlabel('Four separate samples')
# ax[0].set_ylabel('Observed values')

axs[1].violinplot(data[data["Species"] == 1].iloc[:,0:4],showmeans=True, showextrema=True, showmedians=True)
axs[1].set_title('Iris virgicolor')

axs[2].violinplot(data[data["Species"] == 2].iloc[:,0:4],showmeans=True, showextrema=True, showmedians=True)
axs[2].set_title('Iris virginica')
for ax in axs.flat:
    ax.set_xticklabels(["","sl","sw","pl","pw"])
    ax.set_ylabel('Observed values')

fig.suptitle("Violin Plotting Examples")
fig.subplots_adjust(hspace=0.4)
plt.show()
```
